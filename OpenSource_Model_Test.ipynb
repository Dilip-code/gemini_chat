```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "tiiuae/falcon-rw-1b"  # lighter; use tiiuae/falcon-7b-instruct if you have enough GPU
# model_name = "tiiuae/falcon-7b-instruct"  # or falcon-rw-1b , EleutherAI/gpt-neox-20b , Qwen/Qwen3-30B-A3B , tiiuae/Falcon3-10B-Instruct , mistralai/Mistral-Small-3.1-24B-Instruct-2503 , google/gemma-3-27b-it , meta-llama/Llama-4-Scout-17B-16E-Instruct , CohereLabs/c4ai-command-r-plus
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, trust_remote_code=True)

# Move model to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
def chat_with_falcon(prompt, max_new_tokens=200):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=True, temperature=0.7)
    reply = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return reply
```

```python
response = chat_with_falcon("What is the capital of India?.")
print(response)
```
